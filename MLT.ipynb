<h2 align="center">Lecture_6 ML Component Framework in details <br>(Data, Model, Loss Function) </h2>


### Data.

* Training data is represented by **D**.
* There can be many e.g. in the data(**D**).
* Each example may have many features.
* Eacah example is itself a feature vector represented by $X^{(i)}$.
* a general feature vector is represented as <br>
$X^{(i)}$ = [$feature_1, feature_2, feature_3, ...., feature_m $]
* List of label is denoted by **y** in terms of single label, and by **Y** in terms of multi-label.
* **D** is a set of n training e.g.s. a set of ordered pairs (features, labels) represented as **D** = {($X^{(i)}$, $Y^{(i)}$)}
 * Feature matrix represented as **X**. <br>

 **X** = [$X^{(1)^{T}}, X^{(2)^{T}}, X^{(3)^{T}}, X^{(4)^{T}},... ,X^{(n)^{T}}$]
    
>  A vector by default is a column vector.



1. **Dataset Division.**

    * **Training Data.** - For training the model.
    * **Validation Data.** - for tuning any hyper parameters.
    * **Test Data.** - for evaluation of the model performance.
    
    
2. **Cross Validation**, **K-fold Cross Validation.**, **Leave one out CV**.
     
     Instead of using just a fixed set of example for training and validation - differnt sets are used for training and validation.
     

3. **Pre-processing.**

    *  Continous or real value features need to be normalized. for faster convergence.
    * Discrete attributes are converted to numbers by:
        - One-Hot-encoding.
        - hashing.
        - Embeddings.

4. **Relationship b/w model and labels.**

    * Use the histograms, pair plots to check if the correlation exist between the features and labels.
    * few tests like chi-square test.


5. **Handling the missing values.**

    * Remove the egs. containing the missing values.
    * fill the missing values with the mean, mode, median, etc.
    
6. **Class imbalance problem.**

    * Too may eg. from the same class.

<h2 align="center">Lecture_ ML 8 Component Framework in details (Evaluation) </h2>

Model is evaluated on the validation data after the training.

Loss function is a method to evaluation there might be moe methods.

1. **Overfiting and Underfitting.**


|Symptoms|Diagnosis|
|:---:|:---:|
|Training Loss - Low & Validation Loss - Low|Just the right fit|
|Training Loss - high & Validation Loss - high|Underfitting|
|Training Loss - Low & Validation Loss - high|Overfitting|


> Loss V/S Epochs diagram
> Fitting Model.

2. **Fixing Overfiting and Underfitting.**

- Underfitting is caused by the lack of the representation capacity.
    - Adding `polynomial function` can help.
    - A Large `regularizzation` rate can also cause underfitting.
        
        
- Overfitting gis caused by the excessive representation capacity that allows the model to memorize the trainig set and it can't perform well on the unseen examples.
           
    - Supply more data.
    - Introduce the `regularization` to avoide the overfitting, if it has been used, `increase the rate`.

3. **Metrices for Regresion**
    * Mean Squared Error (MSE).
    * Mean Absolute Error.
    
|Metric Name|Mathematical Formula|
|:----:|:----:|
|Mean Squared Error (MSE)|$\frac {1}{n} \sum ^{n} _{i=1} (\^{y}^{i} - y^{i})^2 $|
|Mean Absolute Error.|$\frac {1}{n} \sum ^{n} _{i=1} |(\^{y}^{i} - y^{i})| $|
    
4. **Matrices for Classification.**
    * Precision.
    * Recall.
    * F-1 Score. - Harmonic Mean of the precision and Recall.
    * AUC-ROC. - Area Under the curve - receiver operating charaterstics
    * AUC-PR. - Area Under the curve - Precision Recall Curve

Confusion Matrix:

|Actual\Predicted|False(0)|True(1)|
|:----:|:----:|:----:|
|False(0)|True Negative (TN) |False Positive (FP) |
|True(1)|False Negative(FN)|True Positive(TP) |
    
**Precision** = $\frac {TP}{TP+FP}$ <br>

**Recall** = $\frac {TP}{TP+FN}$ <br>  <br>

**F1** = $2 * \frac {Precision*Recall}{Precision + Recall}$ <br>

**PR Curve**: Precision on Y-Axis and Recall on X=Axis

**ROC Curve**:
    
* **False positive rate : FPR**
        
$FPR = 1 - Specificity = \frac {FP}{FP+TN}$
       
    
* **True Positive Rate : TPR**

$TPR  = Sensitivity = \frac {TP}{FN+TP}$

### Liner Regresion

It is a machine learning model that predicts real valued output label based on `Linear Combination` of `Input Features`.

* Input: $Feature Vector = x_{m * 1}$.
* Label: $ y  \epsilon  R, $ which is `scaler`. (Single lebel set up)

Because it is a model hence it must have the 5 components of ML Framework, namely Data., Model., Loss Function., Optimization., Evaluation.<br>

Let's see how the Data look like in case of the linear regression.<br>

**Data:**
A `set of n ordered pair` of `feature vector`, x and a `label` y representing examples, We denoted it by **D**:

$D$ = ${(X,y)}$ = {$(x^{i}, y^{i})$} $^{n}_{i = 1}$

**Features**:

$X$ is a `feature matrix` corresponding to $n$ `training examples`, each represented with `m features` and has shape $n x m$ . In this matrix, each feature vector is tranposed and represented as a row.


$$ X_{n x m} = \begin{bmatrix} -- (x^{(1)})^T-- \\ -- (x^{(2)})^T-- \\ -- (x^{(3)})^T--\\ .\\. \\. \\ -- (x^{(n)})^T-- \end{bmatrix}$$

feature vector for $i-th$ training example $x^{(i)}$ can be obtained by $X[i]$. slicing in python.

we can conclude that the $x^{i}_{j}$ represent the $j^{th}$ feature of the $i^{th}$ example.

**Labels**

For a single label problem, $y$ is  a labekl vector of shape $n x 1$

$$\begin {bmatrix} y ^{(1)}\\ y ^{(2)}\\ . \\ . \\ . \\y ^{(n)} \end {bmatrix}$$

$i-th$ entry in this vector,$y[i]$ gives label for $i-th$ example, which is denoted by $y^{(i)} \epsilon R $. slicing in python.

**Model**

Label $y$ is computed as a linear combination of feaatures $ x_1, x_2, ..., x_m $:

$$y = w_0 + w_1x_1 +w_2x_2 + .... +w_mx_m $$

here $ w_0 , w_1, w_2, ...., w_m $ are weights.


* Each feature $x_j$ has a corresponding weight $w_j$.
* Let's introduce a `dummy feature` $x_0 = 1$  for weight $w_0$:

$$y = w_0x_0 + w_1x_1 +w_2x_2 + .... +w_mx_m = w^Tx $$

It can now be written compactly as.


$$y = \sum ^m _{i=0}w_ix_i$$

|:----:|:----:|:----:|:----:|

**Loss Function**

$$J(w) = \frac {1}{2} {(Xw-y)^T}{(Xw-y)} $$


<h2 align="center">Lecture_ ML 4 Optimization </h2>

**Optimization**

The key calculation is the derivative of loss function $J(w)$ wit respecct to $w$.<br>

let's look at some usefull identities:

* Derivative of linear function: 
$$\frac {\partial{wx}} {\partial{w}} = \frac {\partial{w^Tx}} {\partial{w}} = \frac {\partial{x^Tw}} {\partial{w}} = x $$

* Derivative of quardetic function:
$$\frac {\partial{w^TX}w} {\partial{w}} = 2Xw $$

* Computing gradient of loss function.
$$\frac {\partial{J(w)}} {\partial{w}} = X^T(Xw-y) $$

where $$J(w) = \frac {1}{2} {(Xw-y)^T}{(Xw-y)} $$

After obtaining the partial derivative how would we get the $w$.

for the same we have.
* **Normal Equation**. - Set the partial derivative to 0 and solve with analytical method to obtain the weight vector.

* **Gradient Descent**. - Iterativly change the weights based on the partial derivative of the loss function untill convergence (iterative optimization)
